# Fashion-MNIST Classification with Neural Networks

This notebook implements various neural network architectures using sklearn to analyze the Fashion-MNIST dataset.

## Data Preparation and Preprocessing

### 1. Data Loading and Visualization
- Loading first 8000 training images and 2000 test images
- Normalizing the data using StandardScaler
- Visualizing 10 random test samples

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
import seaborn as sns

# Load and preprocess data
train_data = pd.read_csv('fashion-mnist_train.csv')
test_data = pd.read_csv('fashion-mnist_test.csv')

# Take first 8000 training samples and 2000 test samples
X_train = train_data.iloc[:8000, 1:].values  # Excluding the label column
y_train = train_data.iloc[:8000, 0].values
X_test = test_data.iloc[:2000, 1:].values
y_test = test_data.iloc[:2000, 0].values

# Normalize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 1. Visualize 10 test samples
plt.figure(figsize=(20, 4))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
    plt.title(f'Label: {y_test[i]}')
plt.show()
```

### 2. MLP Classifier with Different Activation Functions
Training a neural network with architecture [128, 64, 32] using:
- Activation functions: logistic, tanh, relu, identity
- Solver: adam
- Batch size: 128
- Learning rate: 2e-5
- Iterations: 100

The loss curves below show the training progress for each activation function.

```python

# 2. Train MLP Classifier with different activation functions

activations = ['logistic', 'tanh', 'relu', 'identity']
mlp_results = {}

for activation in activations:
    mlp = MLPClassifier(
        hidden_layer_sizes=(128, 64, 32),
        activation=activation,
        solver='adam',
        batch_size=128,
        learning_rate_init=2e-5,
        max_iter=100,
        random_state=42
    )
    mlp.fit(X_train_scaled, y_train)
    
    # Plot loss curves
    plt.figure(figsize=(10, 5))
    plt.plot(mlp.loss_curve_, label='Training loss')
    plt.title(f'Loss Curve - {activation}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    
    mlp_results[activation] = mlp.score(X_test_scaled, y_test)
```

### 3. Hyperparameter Optimization
Performing grid search on the best activation function to optimize:
- Solver: [adam, sgd]
- Learning rate: [1e-5, 2e-5, 1e-4]
- Batch size: [64, 128, 256]

```python

# 3. Grid Search for best hyperparameters
best_activation = max(mlp_results.items(), key=lambda x: x[1])[0]
param_grid = {
    'solver': ['adam', 'sgd'],
    'learning_rate_init': [1e-5, 2e-5, 1e-4],
    'batch_size': [64, 128, 256]
}

grid_search = GridSearchCV(
    MLPClassifier(
        hidden_layer_sizes=(128, 64, 32),
        activation=best_activation,
        max_iter=100,
        random_state=42
    ),
    param_grid,
    cv=3
)
grid_search.fit(X_train_scaled, y_train)

```

```python
# RESULT FOR GRIDSEARCH : #best params : MLPClassifier(activation='tanh', batch_size=64,
#               hidden_layer_sizes=(128, 64, 32), learning_rate_init=2e-05,
#               max_iter=100, random_state=42)

mlp = MLPClassifier(activation='tanh', batch_size=64,
                    hidden_layer_sizes=(128, 64, 32), learning_rate_init=2e-05,
                    max_iter=100, random_state=42)

mlp.fit(X_train_scaled, y_train)
accuracy = mlp.score(X_test_scaled, y_test)
print(f"Accuracy after grid search : {accuracy}")
```

### 4. Image Regeneration Task
Implementing an autoencoder-style network with:
- Architecture: [c, b, a, b, c] where c > b > a
- Two variants: using ReLU and Identity activation functions
- Solver: adam
- Learning rate: 2e-5

Below are the original and regenerated images for comparison.

```python
# Define architecture parameters
c, b, a = 128, 64, 32  

alltrain_features = []
alltest_features = []

for activation in ['relu', 'identity']:
    mlp_reg = MLPRegressor(
        hidden_layer_sizes=(c , b, a, b, c),
        activation=activation,
        solver='adam',
        learning_rate_init=0.001,
        max_iter=100,
        random_state=42
    )
    
    mlp_reg.fit(X_train_scaled, X_train_scaled)
    
    # Visualize regenerated images
    plt.figure(figsize=(20, 8))
    for i in range(10):
        # Original
        plt.subplot(2, 10, i+1)
        plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
        plt.title('Original')
        
        # Regenerated
        regenerated = mlp_reg.predict(X_test_scaled[i:i+1])
        plt.subplot(2, 10, i+11)
        plt.imshow(regenerated.reshape(28, 28), cmap='gray')
        plt.axis('off')
        plt.title('Regenerated')
    plt.suptitle(f'Regeneration with {activation} activation')
    plt.show()
    alltrain_features.append(mlp_reg.predict(X_train_scaled)) 
    alltest_features.append(mlp_reg.predict(X_test_scaled))
```

### 5. Feature Extraction and Classification
Using the bottleneck layer (size 'a') from the regeneration networks as feature extractors:
- Extracting features for train and test sets
- Training smaller MLPs (2 layers of size 'a')
- Comparing classification performance with original MLP classifier

This demonstrates the effectiveness of learned representations for classification.

```python
i = 0
for activation in ['relu', 'identity']:
    
    print(" x shape : ", X_train_scaled.shape)
    # print(" model shape : ", models[i].shape)
    
    # Extract features
    train_features = alltrain_features[i]
    test_features = alltest_features[i]
    
    # train_features = extract_features(mlp_reg, X_train_scaled)
    # test_features = extract_features(mlp_reg, X_test_scaled)
    
    # Print shapes to verify
    print(f"Train features shape: {train_features.shape}")
    print(f"Train labels shape: {y_train.shape}")
    print(f"Test features shape: {test_features.shape}")
    print(f"Test labels shape: {y_test.shape}")
    
    # Train smaller MLP Classifier
    small_mlp = MLPClassifier(
        hidden_layer_sizes=(a, a),
        activation=activation,
        solver='adam',
        learning_rate_init=2e-5,
        max_iter=200,
        random_state=42
    )
    small_mlp.fit(train_features, y_train)
    
    print(f"Accuracy with {activation} features:", small_mlp.score(test_features, y_test))
    i += 1
```

### Results and Analysis
Key findings:

3. Feature extraction classification performance :
